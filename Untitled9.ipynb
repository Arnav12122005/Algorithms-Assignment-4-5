```python

import time, random
from typing import List, Tuple, Dict, Optional
import matplotlib.pyplot as plt

Flight = Tuple[str, int, int]
Assignment = Dict[str, List[str]]


def print_assign(assign: Assignment):
    if not assign:
        print("No assignment (None).")
        return
    for c, fl in assign.items():
        print(f"{c}: {', '.join(fl) if fl else 'â€”'}")


def sample_flights(n:int, start_hour:int=6, end_hour:int=22, min_len:int=1, max_len:int=3, seed:Optional[int]=None) -> List[Flight]:
    """Generate n random flights (id, start, end)."""
    if seed is not None:
        random.seed(seed)
    flights=[]
    for i in range(n):
        dur = random.randint(min_len, max_len)
        s = random.randint(start_hour, end_hour - dur)
        e = s+dur
        flights.append((f"F{i+1}", s, e))
    random.shuffle(flights)
    return flights

def plot_gantt(assign_map:Assignment, flights:List[Flight], title="Gantt Chart"):
    """Simple matplotlib gantt for visualization."""
    fmap = {fid:(s,e) for fid,s,e in flights}
    crews = list(assign_map.keys())
    fig, ax = plt.subplots(figsize=(10, 0.8*len(crews)+1.5))
    for i,c in enumerate(crews):
        y = i
        for fid in assign_map[c]:
            s,e = fmap[fid]
            ax.broken_barh([(s, e-s)], (y-0.3, 0.6), alpha=0.7)
            ax.text(s+0.05, y, fid, va='center', fontsize=9)
    ax.set_yticks(range(len(crews)))
    ax.set_yticklabels(crews)
    ax.set_xlabel("Time (hours)")
    ax.set_title(title)
    ax.set_xlim(0,24)
    plt.grid(True, axis='x', linestyle='--', alpha=0.3)
    plt.show()




def can_assign(schedule_for_crew:List[Flight], flight:Flight, min_rest:int=1) -> bool:
    """Return True if 'flight' can be added to schedule_for_crew with min_rest hours."""
    fid,s,e = flight
    for _, s2, e2 in schedule_for_crew:

        if not (e + min_rest <= s2 or e2 + min_rest <= s):
            return False
    return True


_demo_schedule = [('F1', 9, 11), ('F3', 13, 15)]
_demo_new = ('F2', 11, 12)
print("Part 2 demo can_assign:", can_assign(_demo_schedule, _demo_new, min_rest=1))



def backtrack_assign(flights:List[Flight], crew:List[str], min_rest:int=1, time_limit:Optional[float]=None):
    """
    Backtracking solver:
      - tries to assign all flights to crew members
      - returns the first found full assignment (or None if none found within time_limit)
    """
    start_time = time.time()
    flights_sorted = sorted(flights, key=lambda x:x[1])
    n = len(flights_sorted)
    schedules = {c:[] for c in crew}
    best = {'assign': None}
    stats = {'calls':0}
    deadline = start_time + time_limit if time_limit else None

    def rec(i):
        if deadline and time.time() > deadline:
            return
        stats['calls'] += 1
        if i == n:
            best['assign'] = {c: [f[0] for f in schedules[c]] for c in crew}
            return
        flight = flights_sorted[i]
        for c in crew:
            if can_assign(schedules[c], flight, min_rest):
                schedules[c].append(flight)
                rec(i+1)
                schedules[c].pop()
                if best['assign'] is not None:
                    return
    rec(0)
    return best['assign'], stats


flights_demo = [('F1', 9,11), ('F2',10,12), ('F3',13,15), ('F4',11,12)]
crew_demo = ['C1','C2','C3']
assign, stats = backtrack_assign(flights_demo, crew_demo, min_rest=1, time_limit=1.0)
print("Part 3 assignment:")
print_assign(assign)
print("Calls:", stats['calls'])



def greedy_earliest_finish(flights:List[Flight], crew:List[str], min_rest:int=1):
    """Sort flights by end time and assign to first compatible crew."""
    flights_sorted = sorted(flights, key=lambda x:x[2])
    schedules = {c:[] for c in crew}
    for f in flights_sorted:
        for c in crew:
            if can_assign(schedules[c], f, min_rest):
                schedules[c].append(f)
                break
    return {c: [fl[0] for fl in schedules[c]] for c in crew}

def greedy_least_loaded(flights:List[Flight], crew:List[str], min_rest:int=1):
    """Assign flights in start-time order to the crew with fewest flights that can take it."""
    flights_sorted = sorted(flights, key=lambda x:x[1])
    schedules = {c:[] for c in crew}
    for f in flights_sorted:

        crews_sorted = sorted(crew, key=lambda c: len(schedules[c]))
        for c in crews_sorted:
            if can_assign(schedules[c], f, min_rest):
                schedules[c].append(f)
                break
    return {c: [fl[0] for fl in schedules[c]] for c in crew}


fl = sample_flights(8, seed=42)
cr = ['C1','C2','C3']
print("Flights:", fl)
g1 = greedy_earliest_finish(fl, cr)
g2 = greedy_least_loaded(fl, cr)
print("Greedy (earliest finish):"); print_assign(g1)
print("Greedy (least loaded):"); print_assign(g2)
plot_gantt(g1, fl, title="Greedy: Earliest Finish")

def scaling_experiment(max_n:int=10, crew_size:int=3, repeats:int=3, time_limit:float=0.5):
    out=[]
    for n in range(2, max_n+1):
        calls_sum=0
        found = 0
        for r in range(repeats):
            fl = sample_flights(n, seed=1000 + n*10 + r)
            crew = [f"C{i+1}" for i in range(crew_size)]
            assign, stats = backtrack_assign(fl, crew, min_rest=1, time_limit=time_limit)
            if assign: found+=1
            calls_sum += stats['calls']
        out.append((n, found, calls_sum//repeats))
    return out

print("Part 5 scaling experiment (example):")
print(scaling_experiment(max_n=8, crew_size=3, repeats=5, time_limit=0.25))

def make_viva_notes_partwise():
    notes = {
        1: "Data generation: deterministic seed used for repeatability.",
        2: "Feasibility: checks for overlap + configurable minimum rest (safety constraint).",
        3: "Exact solver: recursive backtracking to find full feasible assignment; time-limited for practicality.",
        4: "Heuristics: earliest-finish and least-loaded for near-instant results on large inputs.",
        5: "Experiments: time-limited runs show the solver scales poorly; heuristics scale well.",
        6: "Visualization: Gantt chart to present schedules in viva; easy to explain crew timelines."
    }
    return notes

print("\nViva ready notes (short):")
for k,v in make_viva_notes_partwise().items():
    print(f"Part {k}: {v}")

report_text = "Sample run:\n\nFlights: " + str(fl) + "\n\nGreedy assignment:\n" + str(g1)
with open("/content/sample_scheduler_report.txt","w") as f:
    f.write(report_text)
print("\nSaved /content/sample_scheduler_report.txt")


```
